```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
plt.style.use('ggplot')
np.random.seed(123)
```

```python
p_trans = np.zeros([3, 3])
```

```python
# 0: office 1:kyuukei 2:jikken
p_trans[0, 0] = 0.1
p_trans[0, 1] = 0.2
p_trans[0, 2] = 0.7
p_trans[1, 0] = 0.1
p_trans[1, 1] = 0.4
p_trans[1, 2] = 0.5
p_trans[2, 0] = 0.3
p_trans[2, 1] = 0.3
p_trans[2, 2] = 0.4
```

```python
p_trans
```

```python
p_trans[0]
```

```python
c_state = 0
np.random.choice(3, 1, p=p_trans[c_state])
```

```python
NMCS = 400
c_state = 0 # office
c_arr = [c_state]
for _ in range(NMCS):
    current = np.random.choice(3, 1, p=p_trans[c_state])
    c_state = current[0]
    c_arr.append(c_state)
df = pd.DataFrame(c_arr)
```

```python
df.head()
```

```python
plt.plot(df[0])
plt.xlabel('MCS')
plt.ylabel('place')
```

```python
#lt.hist([df[:10], df[:50], df[:200], df[:400]], normed=True, label=["10", "50", "200", "400"])
plt.hist([df[0][:10], df[0][:50], df[0][:200], df[0][:400]], normed=True, label=["10", "50", "200", "400"])
plt.xlabel('place')
plt.ylabel('frequency')
plt.legend()
```
