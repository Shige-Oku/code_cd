## Trainデータ編集

```python
train_cp = train_pd.copy()
```

```python
train_cp.drop(['amenities', 'cancellation_policy', 'description', 'latitude', 'longitude', 'name', 'neighbourhood', 'property_type', 'thumbnail_url', 'zipcode', 'bed_type', 'city', 'first_review', 'host_since', 'last_review', 'room_type', 'host_response_rate'], axis=1, inplace=True)
```

```python
train_cp['cleaning_fee'][train_cp['cleaning_fee'] == 't'] = 0
```

```python
train_cp['cleaning_fee'][train_cp['cleaning_fee'] == 'f'] = 1
```

```python
train_cp.cleaning_fee.value_counts()
```

```python
train_cp['host_has_profile_pic'][train_cp['host_has_profile_pic'] == 't'] = 0
```

```python
train_cp['host_has_profile_pic'][train_cp['host_has_profile_pic'] == 'f'] = 1
```

```python
train_cp.host_has_profile_pic.value_counts()
```

```python
train_cp['host_identity_verified'][train_cp['host_identity_verified'] == 't'] = 0
```

```python
train_cp['host_identity_verified'][train_cp['host_identity_verified'] == 'f'] = 1
```

```python
train_cp.host_identity_verified.value_counts()
```

```python
train_cp['instant_bookable'][train_cp['instant_bookable'] == 't'] = 0
```

```python
train_cp['instant_bookable'][train_cp['instant_bookable'] == 'f'] = 1
```

```python
train_cp.instant_bookable.value_counts()
```

```python
# train_pd[train_pd['host_response_rate'].isnull()] = '0%'
```

```python
# train_pd['host_response_rate'] = train_pd.host_response_rate.apply(lambda x: re.sub(r'\D', '', x))
# train_pd['host_response_rate'] = train_pd.host_response_rate.apply(lambda x: x + '%')
```

```python
train_cp.head()
```

```python
train_cp = train_cp.dropna(subset=['bathrooms', 'bedrooms', 'beds', 'host_has_profile_pic', 'host_identity_verified'])
```

```python
train_cp = train_cp.fillna({'review_scores_rating' : train_cp.review_scores_rating.median()})
```

```python
chv.check_miss_value(train_cp)
```

```python
train_cp[['accommodates', 'bathrooms', 'bedrooms', 'beds', 'cleaning_fee', 'host_has_profile_pic', 'host_identity_verified', 'instant_bookable', 'review_scores_rating']] = train_cp[['accommodates', 'bathrooms', 'bedrooms', 'beds', 'cleaning_fee', 'host_has_profile_pic', 'host_identity_verified', 'instant_bookable', 'review_scores_rating']].astype('int8')
```

```python
train_cp['number_of_reviews'] = train_cp['number_of_reviews'].astype('int16')
```
