```python
# 0 - 25 の範囲で等間隔に 100個
x = np.linspace(0, 25, 100)
```

```python
x
```

```python
y1 = np.sin(x)
```

```python
y2 = np.cos(x)
```

```python
y1
```

```python
y2
```

```python
plt.plot(x, y1, '-b', label='sine')
plt.plot(x, y2, '-r', label='cosine')
plt.legend(loc='upper left')
plt.ylim(-1.5, 2.0)
plt.show()
```

```python
plt.plot(x, y1, '-b', label='sine')
plt.plot(x, y2, '-r', label='cosine')
plt.legend(loc='upper left')
plt.title("Sine & Cosine plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.show()
```

```python
def sq(x):
    return x ** 2
```

```python
def cube(x):
    return x ** 3
```

```python
def power4(x):
    return x ** 4
```

```python
plt.plot(x, sq(x), 'r')
plt.plot(x, cube(x), 'b')
plt.plot(x, power4(x), 'g')
plt.show()
```

```python
plt.plot(x, sq(x), 'r')
plt.plot(x, cube(x), 'b')
plt.plot(x, power4(x), 'g')
plt.axis([0, 5, 0, 200])
plt.show()
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
# plt.subplots(ncols=2) 2列のサブプロットの作成
```

```python
v = np.linspace(-np.pi, np.pi, 100)
```

```python
v
```

```python
x1 = np.sin(2 * v)
```

```python
x2 = np.cos(2 * v)
```

```python
x1
```

```python
x2
```

```python
figure, (Left, Right) = plt.subplots(ncols=2, figsize=(10, 10))
# 左側のグラフの設定
Left.plot(v, x1, linewidth=2)
Left.set_title('sin')
Left.set_xlabel('values')
Left.set_ylabel('x')
Left.set_xlim(-np.pi, np.pi)
Left.grid(True)
# 右側のグラフの設定
Right.plot(v, x2, linewidth=2)
Right.set_title('cos')
Right.set_xlabel('values')
Right.set_ylabel('x')
Right.set_xlim(-np.pi, np.pi)
Right.grid(True)
# 両方のグラフの設定
figure.show()
```

```python
# chapter 8 ヒストグラム：棒グラフ　横軸：範囲、階級　縦軸：個数　
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 標準正規分布 = Standard Normal Distribution
# 標準正規分布は平均が0、分散が1の正規分布
# 連続一様分布 = Continuous Uniform Distribution
# np.random.randn() 標準正規分布に属する乱数生成
# np.random.random() 連続一様分布に属する乱数生成
# plt.hist() ヒストグラムの作成
```

```python
# 標準正規分布は平均が0、分散が1の正規分布
g = np.random.randn(1000)
```

```python
g.mean()
```

```python
g.std()
```

```python
plt.hist(g)
plt.show()
```

```python
plt.hist(g, bins=100)
```

```python
# range：範囲
plt.hist(g, bins=100, range=(0, 5))
```

```python
# cumulative：累積値
plt.hist(g, bins=100, cumulative=True)
plt.show()
```

```python
# rwidth：幅　histtype：塗りつぶし
plt.hist(g, bins=100, rwidth=0.5, color='red', histtype="step")
plt.show()
```

```python
# ２つの表を表示
# 連続一様分布
n = np.random.random(1000)
```

```python
n
```

```python
labels = ['Gaussian', 'CUD']
```

```python
# 正規分布から設定
# alpha：透明度
plt.hist(g, bins=100, alpha=0.3, color="red")
plt.hist(n, bins=100, alpha=0.3, color="blue")
plt.show()
```
