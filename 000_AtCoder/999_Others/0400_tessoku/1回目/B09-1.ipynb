```python
N = int(input())
```

```python
ABCDn = [list(map(int, input().split())) for _ in range(N)]
```

```python
pa = [[0] * 1502 for _ in range(1502)]
for A, B, C, D in ABCDn:
    pa[A][B] += 1
    pa[A][D] -= 1
    pa[C][B] -= 1
    pa[C][D] += 1
for h in range(1502):
    for w in range(1502):
        pa[h][w] += pa[h][w-1]
for w in range(1502):
    for h in range(1502):
        pa[h][w] += pa[h-1][w]
        
```

```python
ans = 0
for p in pa:
    ans += 1502 - p.count(0)
print(ans)
```
