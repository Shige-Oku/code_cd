```python
H, W = map(int, input().split())
```

```python
Cn = ['#' + input() + '#' for _ in range(H)]
```

```python
Cn.insert(0, '#' * (W+2))
Cn.append('#' * (W+2))
```

```python
def DFS(s, g, y, x, c, dis=None):
    print(s, g, y, x, c)
    ret = -1
    if dis is None:
        dis = [[-1] * (W+2) for _ in range(H+2)]
    dis[y][x] = c
    for y2, x2 in ((-1, 0), (1, 0), (0, 1), (0, -1)):
        if y+y2 == y & x+x2 == x:
            return dis[y][x] + 1
        if Cn[y+y2][x+x2] == '#':
            continue
        if dis[y+y2][x+x2] != -1:
            continue
        ret2 = DFS(s, g, y+y2, x+x2, c+1, dis)
        if ret2 != -1:
            ret = max(ret, ret2)
    dis[y][x] = -1
    print(ret)
    print(dis)
    return ret
```

```python
ans = -1
for i in range(H):
    for j in range(W):
        if Cn[i+1][j+1] == '#':
            continue
        ans = max(ans, DFS(i+1, j+1, i+1, j+1, 0))
```
