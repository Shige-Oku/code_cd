```python
N, A, B, C = map(int, input().split())
```

```python
Ln = [int(input()) for _ in range(N)]
```

```python
def DFS(cur, a, b, c):
    if cur >= N:
        return abs(a-A) + abs(b-B) + abs(c-C) - 30 if min(a, b, c) > 0 else 10**10
    ret1 = DFS(cur+1, a, b, c)
    ret2 = DFS(cur+1, a+Ln[cur], b, c) + 10
    ret3 = DFS(cur+1, a, b+Ln[cur], c) + 10
    ret4 = DFS(cur+1, a, b, c+Ln[cur]) + 10
    return min(ret1, ret2, ret3, ret4)
```

```python
import heapq
from itertools import product
def solve(P):
    if sum(P) < 3:
        return 10**10
    ret = 0
    q = []
    leng = []
    for i, p in enumerate(P):
        if p == 1:
            leng.append(Ln[i])
    for cp in product([0, 1], repeat=len(leng)):
        if sum(cp) == 0:
            continue
        if sum(cp) == 1:
            for i, p in enumerate(cp):
                if p == 1:
                    heapq.heappush(q, (0, leng[i]))
                    break
        else:
            l = 0
            for i, p in enumerate(cp):
                if p == 0:
                    continue
                l += leng[i]
            else:
                heapq.heappush(q, ((sum(cp)-1)*10, l))
    flg = [False] * 3
    while q:
        if flg[0] and flg[1] and flg[2]:
            break
        c, l = heapq.heappop(q)
        print(c, l, flg)
        if not flg[0]:
            if l == A:
                ret += c
                flg[0] = True
                continue
        if not flg[1]:
            if l == B:
                ret += c
                flg[1] = True
                continue
        if not flg[2]:
            if l == C:
                ret += c
                flg[2] = True
                continue
        heapq.heappush(q, (c+1, l+1))
        if l > 2:
            heapq.heappush(q, (c+1, l-1))
    return ret
```

```python
from itertools import product
ans = 10**10
for P in product([0, 1], repeat=N):
    print(P)
    c = solve(P)
    ans = min(ans, c)
    print(ans)
```
