```python
N, A, B, C, D = map(int, input().split())
```

```python
S = input()
```

```python
SS = "#" + S + "##"
flg1 = True
flg2 = True
if C < D:
    flg3 = True
else:
    flg3 = False
```

```python
for i in range(1, N+1):
    if A <= i <= C-1:
        if SS[i] == SS[i+1] == "#":
            flg1 = False
    if B <= i <= D-1:
        if SS[i] == SS[i+1] == "#":
            flg2 = False

for i in range(B-1, D):
    if SS[i] == SS[i+1] == SS[i+2] == '.':
        flg3 = True
```

```python
if flg1 and flg2 and flg3:
    print("Yes")
else:
    print("No")
```
