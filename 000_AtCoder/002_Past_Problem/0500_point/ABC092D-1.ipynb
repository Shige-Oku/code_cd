```python
grid = []
for _ in range(50):
    grid.append(['.'] * 50)
for _ in range(50):
    grid.append(['#'] * 50)
```

```python
cnt = 1
for i in range(0, 50, 2):
    for j in range(0, 50, 2):
        if cnt >= A:
            break
        grid[i][j] = '#'
        cnt += 1
    if cnt >= A:
        break
cnt = 1
for i in range(51, 100, 2):
    for j in range(0, 50, 2):
        if cnt >= B:
            break
        grid[i][j] = '.'
        cnt += 1
    if cnt >= B:
        break
        
```
