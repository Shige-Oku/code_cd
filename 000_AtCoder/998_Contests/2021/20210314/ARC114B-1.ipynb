```python
import sys
import numpy as np
import math
import collections
import copy
import decimal
import itertools
import re
import bisect
import scipy
from collections import deque 
from functools import reduce
from itertools import product
from itertools import combinations
from functools import lru_cache
from scipy.special import comb
from numba import njit
from bisect import bisect_right
from bisect import bisect_left
```

```python
N = int(input())
```

```python
Fn = list(map(int, input().split()))
```

```python
graph = [[] for _ in range(N+1)]
for i, F in enumerate(Fn, 1):
    graph[i].append(F)
```

```python
visit = [False for _ in range(N+1)]
rt = 0
i = 1
while i <= N:
    if visit[i]:
        i += 1
        continue
    if visit[Fn[i-1]]:
        visit[i] = True
        i += 1
        continue
    rt += 1
    q = deque()
    q.append(i)
    while q:
        v = q.popleft()
        if visit[v]:
            continue
        visit[v] = True
        for gr in graph[v]:
            if visit[gr]:
                continue
            q.append(gr)
    i += 1
    
```
