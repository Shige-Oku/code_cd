```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, norm

# 正規分布のデータ生成
data_normal = np.random.normal(0, 1, 1000)
kurt_normal = kurtosis(data_normal)

# 低尖度分布のデータ生成
data_flat = np.random.uniform(-1, 1, 1000)
kurt_flat = kurtosis(data_flat)

# 高尖度分布のデータ生成
data_peak = np.random.normal(0, 0.5, 1000) + np.random.normal(2, 0.5, 1000)
kurt_peak = kurtosis(data_peak)

# グラフ化
fig, axs = plt.subplots(3, 1, figsize=(6, 9))

axs[0].hist(data_normal, bins=50)
axs[0].set_title(f"Normal Distribution (kurtosis={kurt_normal:.2f})")

axs[1].hist(data_flat, bins=50)
axs[1].set_title(f"Flat Distribution (kurtosis={kurt_flat:.2f})")

axs[2].hist(data_peak, bins=50)
axs[2].set_title(f"Peak Distribution (kurtosis={kurt_peak:.2f})")

plt.tight_layout()
plt.show()
```
