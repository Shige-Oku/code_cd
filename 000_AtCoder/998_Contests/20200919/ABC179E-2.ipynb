```python
N, X, M = map(int, input().split())
```

```python
def f(X, M):
    if X**2 < M:
        return X**2
    else:
        return X**2 % M
```

```python
idx = np.full(M, -1)
loop_st = -1
An = []
val = X
An.append(val)
idx[X] = 0
for i in range(1, N):
    val = f(val, M)
#     Pdb().set_trace()
    if (idx[val] != -1):
        loop_st = idx[val]
#         print('break')
#         print(i, idx[val], loop_st)
        break
    An.append(val)
    idx[val] = i
#     print(i, val, idx[val])
```

```python
An.insert(0, 0)
cum = np.cumsum(An)
```

```python
ans = 0
if N <= len(An):
    print(cum[N])
else:
    lp_b = cum[loop_st-1]
    lp_cnt = i - loop_st
    qu = (N-loop_st) // lp_cnt
    lp_d = (cum[-1] - cum[loop_st]) * qu
    mo = (N-loop_st) % lp_cnt
    if mo == 0:
        lp_a = 0
    else:
        lp_a = cum[loop_st+mo-1] - cum[loop_st-1]
    print(lp_b + lp_d + lp_a)
```
