```python
def DFS(root, seen, vx, vy):
#     print(vx, vy)
    seen[vx][vy] = True
    for x, y in ((1, 0), (-1, 0), (0, 1), (0, -1)):
#         if vx + x < 0 or vx + x > H or vy + y < 0 or vy + y > W:
#              continue
        if (root[vx+x][vy+y] == "#" or seen[vx+x][vy+y]):
            continue
        DFS(root, seen, vx+x, vy+y)
    return
        
```
