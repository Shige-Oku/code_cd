```python
import cv2

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# カラー画像読み込み
logo = cv2.imread('logo_cv2.png', 1)
```

```python
logo.shape
```

```python
# OpenCVのimreadでカラー画像データを読み込む場合、初期設定はBGR（Blue、Green、Red0）
# cv2.cvtColor()：順序をRGB（Red、Green、Blue）へ変更する
# BGR => RGB
logo = cv2.cvtColor(logo, cv2.COLOR_BGR2RGB)
```

```python
logo.shape
```

```python
# Red
red = cv2.imread('logo_cv2.png', 1)
red[:, :, 1] = 0
red[:, :, 2] = 0
```

```python
# Green
green =  cv2.imread('logo_cv2.png', 1)
green[:, :, 0] = 0
green[:, :, 2] = 0
```

```python
# Blue
blue =  cv2.imread('logo_cv2.png', 1)
blue[:, :, 0] = 0
blue[:, :, 1] = 0
```

```python
fig, ax = plt.subplots(1, 4, figsize=(6, 3))

# オリジナル
plt.subplot(141)
plt.imshow(logo)
plt.title('original')

# Red
plt.subplot(142)
plt.imshow(red)
plt.title('red')

# Green
plt.subplot(143)
plt.imshow(green)
plt.title('green')

# Blue
plt.subplot(144)
plt.imshow(blue)
plt.title('blue')
```

```python
# cat
cat = cv2.imread('cat.jpg', 1)
```

```python
cat.shape
```

```python
# ぼかす
cat_blur1 = cv2.blur(cat, (5, 5))

cat_blur2 = cv2.blur(cat, (50, 50))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 20))

# オリジナル
plt.subplot(131)
plt.imshow(cat)
plt.title('Original')

# ぼかし 5x5
plt.subplot(132)
plt.imshow(cat_blur1)
plt.title('Blur 5x5')

# ぼかし 50x50
plt.subplot(133)
plt.imshow(cat_blur2)
plt.title('Blur 50x50')
```

```python
# BGR => RGB
cat_rgb = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)
cat_rgb_blur1 = cv2.cvtColor(cat_blur1, cv2.COLOR_BGR2RGB)
cat_rgb_blur2 = cv2.cvtColor(cat_blur2, cv2.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 20))

# オリジナル
plt.subplot(131)
plt.imshow(cat_rgb)
plt.title('Original')

# ぼかし 5x5
plt.subplot(132)
plt.imshow(cat_rgb_blur1)
plt.title('Blur 5x5')

# ぼかし 50x50
plt.subplot(133)
plt.imshow(cat_rgb_blur2)
plt.title('Blur 50x50')
```

```python
cat2 = cv2.imread('cat.jpg', 1)
cat2_rgb = cv2.cvtColor(cat2, cv2.COLOR_BGR2RGB)

plt.imshow(cat2_rgb)
```

```python
# エッジ検出（閾値 5, 50）
cat2_edge1 = cv2.Canny(cat2_rgb, 5, 50)

plt.imshow(cat2_edge1)
```

```python
# エッジ検出（閾値 100, 200）
cat2_edge2 = cv2.Canny(cat2_rgb, 100, 200)

plt.imshow(cat2_edge2)
```

```python
cat3 = cv2.imread('cat.jpg', 1)
cat3_rgb = cv2.cvtColor(cat3, cv2.COLOR_BGR2RGB)

plt.imshow(cat3_rgb)
```

```python
cat3_rgb.shape
```

```python
# 猫の目の部分を抽出
cat3_eye = cat3_rgb[220:300, 180:350, :]

plt.imshow(cat3_eye)
```

```python
# 画像に丸の書き込み
cat4 = cv2.imread('cat.jpg', 1)
cat4_rgb = cv2.cvtColor(cat4, cv2.COLOR_BGR2RGB)
cat4_circle = cv2.circle(cat4_rgb, (240, 340), 50, color=(255, 0, 0), thickness=3)

plt.imshow(cat4_circle)
```

```python
# 画像に文字を書き込み
cat5 = cv2.imread('cat.jpg', 1)
cat5_rgb = cv2.cvtColor(cat5, cv2.COLOR_BGR2RGB)
cat5_text = cv2.putText(
     cat5_rgb, "codexa", org=(300, 100),
     fontFace=cv2.FONT_HERSHEY_SIMPLEX,
     fontScale=2,
     color=(0, 150, 0), thickness=5
)

plt.imshow(cat5_text)
```
